```python
#BipinPoudel ACE077BCT029
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/drive/MyDrive/pythonlab/harry_potter_reviews.csv")
df.head(5)
```

```python
# Categorical data
# countplot
sns.countplot(data=df, x='user_sex')
df['user_age'].value_counts()
```

```python
 # pie chart

df['user_sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['user_age'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['user_age'])
```

```python
# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['user_age'])
```

```python
df.describe()
```

```python
# Scatter plot(Numerical - Numerical) for the relation between two column
plt.figure(figsize=(17, 7))
sns.scatterplot(x='user_age', y='user_country',hue='user_sex', style='favourite_character',size='rating' ,data=df)
plt.show()
```

```python
 # Bar plot (Numerical - Categorical)
  # average age in pclass
plt.figure(figsize=(17, 7))
sns.barplot(x='user_country',y='user_age',hue='user_sex',data=df)
```

```python

#Box plot (Numerical - Categorical)
plt.figure(figsize=(10, 7))
sns.boxplot(x='user_sex',y='user_age',hue='rating',data=df)
```

```python
# HeatMap (Categorical - Categorical)
sns.heatmap(pd.crosstab(df['user_sex'],df['rating']))
```

```python

# pairplot
plt.figure(figsize=(10, 7))
sns.pairplot(df,hue='user_sex')
```

```python
plt.figure(figsize=(10, 7))
sns.lineplot(x='user_age',y='favourite_character',data=df)
```
